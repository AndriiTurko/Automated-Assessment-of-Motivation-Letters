```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/letters_with_total_score.csv")
```

```python
print(f"Total letters: {len(df)}")
print(df.describe())
```

```python
# Word Count Analysis
plt.figure(figsize=(10,6))
sns.histplot(df['word_count'], bins=30, kde=True)
plt.title('Distribution of Word Counts')
plt.xlabel('Word Count')
plt.ylabel('Number of Letters')
# add vertical lines at 800 and 900
plt.axvline(x=800, color='r', linestyle='-', label='800 words')
plt.axvline(x=900, color='r', linestyle='-', label='900 words')
plt.grid(True)
plt.show()
```

```python
# Grammar Errors Analysis
plt.figure(figsize=(10,6))
sns.histplot(df['grammar_errors'], bins=30, kde=True)
plt.title('Distribution of Grammar Errors')
plt.xlabel('Number of Grammar Errors')
plt.ylabel('Number of Letters')
plt.grid(True)
plt.show()
```

```python
# Total Score Analysis
plt.figure(figsize=(10,6))
sns.histplot(df['total_score'], bins=20, kde=True)
plt.title('Distribution of Total Scores')
plt.xlabel('Total Score (out of 45)')
plt.ylabel('Number of Letters')
plt.grid(True)
plt.show()
```

```python
# Correlation between grammar errors and total score
plt.figure(figsize=(10,6))
sns.scatterplot(x='grammar_errors', y='total_score', data=df)
plt.title('Grammar Errors vs Total Score')
plt.xlabel('Grammar Errors')
plt.ylabel('Total Score')
plt.grid(True)
plt.show()
```

```python
# Correlation between word count and total score
plt.figure(figsize=(10,6))
sns.scatterplot(x='word_count', y='total_score', data=df)
plt.title('Word Count vs Total Score')
plt.xlabel('Word Count')
plt.ylabel('Total Score')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('data/letters_with_grammar.csv')
```

```python
# get max value of questions columns
question_cols = [col for col in df.columns if col.startswith("question_")]
question_cols
```

```python
df[question_cols].max()
```

```python
df[question_cols].min()
```

```python
df[df['question_7'] < 0.01]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in question_cols:
    plt.figure()
    sns.histplot(df[col], bins=50, binrange=(0, 1))
    plt.title(f"Distribution for {col}")
    plt.xlabel("Score")
    plt.xticks([round(x * 0.1, 1) for x in range(11)])  # 0.0 to 1.0
    plt.ylabel("Count")
    plt.grid(True)
    plt.show()
```
